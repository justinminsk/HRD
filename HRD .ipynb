```python
import pandas as pd
import pygal
import numpy as np
#import libraries
```

```python
HRD = pd.read_csv('HR_comma_sep.csv')
HRD.head()
#make a varible for the data and show it
```

```python
HRD.columns
#show column names
```

```python
count_salary = pd.value_counts(HRD['salary'], sort = True).sort_index()
count_salary
#how many of each salary level occur
```

```python
count_sales = pd.value_counts(HRD['sales'], sort = True).sort_index()
count_sales
#how many of each job title occur
```

```python
count_tsc = pd.value_counts(HRD['time_spend_company'], sort = True).sort_index()
count_tsc
#how many people work at each year point
```

```python
HRD.describe()
#summary data
```

```python
HRD.groupby('salary').mean()
#means of each number column broken down by salary level
```

```python
HRD.groupby('time_spend_company').mean()
#means of each number column broken down by how mayn years they have been in the company
```

```python
HRD.groupby('sales').mean()
#means of each number column broken down into job title
```

```python
bar = pygal.Bar()
bar.title = 'Time spent in company Histogram'
bar.add('Frequency', count_tsc)
bar.x_labels = map (str,(2, 3, 4, 5, 6, 7, 8, 10))
bar.render_to_file('histogram_tsc.svg')
#bar graph of the time spent in company broken by how many of each occurance
```

```python
bar = pygal.Bar()
bar.title = 'Job in company Histogram'
bar.add('Frequency', count_sales)
bar.x_labels = map (str, ('IT', 'RandD', 'accounting', 'hr', 'management', 'marketing', 'product_mng', 'sales', 'support', 'technical'))
bar.render_to_file('histogram_sales.svg')
#bar graph of the job title in the company broken by how many of each occurance
```

```python
bar = pygal.Bar()
bar.title = 'Salary in company Histogram'
bar.add('Frequency', count_salary)
bar.x_labels = map (str,(2, 3, 4, 5, 6, 7, 8, 10))
bar.render_to_file('histogram_salary.svg')
#bar graph of the salary breakdown broken by how many of each occurance
```

```python
HRD.groupby(['salary', 'sales']).mean()
#break down of the data by salary and job titles then taking the mean of each number column
```
